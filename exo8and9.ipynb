```python
#excercice 9 
def first_n_primes(N):
    if N < 1:
        return []
    primes = []
    size = max(15, N * 10)  
    while True:
        sieve = [True] * size
        sieve[0] = sieve[1] = False 
        for i in range(2, int(size**0.5) + 1):
            if sieve[i]:
                for j in range(i*i, size, i):
                    sieve[j] = False
        primes = [i for i, is_prime in enumerate(sieve) if is_prime]
        if len(primes) >= N:
            return primes[:N]    
print(first_n_primes(10))
```
