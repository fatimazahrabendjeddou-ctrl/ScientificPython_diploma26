```python
#excercice 1
def solve_quadratic_eqt(a, b, c):
    if a == 0:
        print("This is not a quadratic equation (a = 0).")
        return
    delta = b**2 - 4*a*c
    if delta < 0:
        print("No real solutions because delta<0")
    elif delta == 0:
        x = -b / (2*a)
        print("One real solution which is :", x)
    else:
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        print("The solutions are :", x1, "and", x2)
```

```python
solve_quadratic_eqt(1, -3, 2)
```

```python
#excerice 2 
def recaman(N):
    seq = [0]       
    seet = {0}       
    for n in range(1, N):
        c = seq[n-1] - n   
        if c > 0 and c not in seet:
            seq.append(c)  
            seet.add(c)          
        else:
            new_num = seq[n-1] + n     
            seq.append(new_num)  
            seet.add(new_num)         
    return seq
```

```python
print(recaman(48))
```

```python
#excercice 5 
def pairs(n):
    p = []
    for i in range(1, int(n**0.5) + 1):  
        if n % i == 0:
            p.append((i, n // i))  
    return p
```

```python
print(pairs(12))
```

```python
import time
REPEAT = 10000
# 1
start = time.time()
for _ in range(REPEAT):
    lst = []
    for i in range(1000):
        if i % 3 == 0:
            lst.append(i)
end = time.time()
print("Naive time:", end - start)
# 2
start = time.time()
for _ in range(REPEAT):
    lst = [i for i in range(1000) if i % 3 == 0]
end = time.time()
print("List comprehension time:", end - start)
# 3
start = time.time()
for _ in range(REPEAT):
    lst = list(range(1000))[::3]
end = time.time()
print("Slicing time:", end - start)
# 4
start = time.time()
for _ in range(REPEAT):
    lst = list(range(0, 1000, 3))
end = time.time()
print("Range time:", end - start)
```

```python
#excercice 7 
def is_palindrome(s):
    s = s.lower()
    return s == s[::-1]
```

```python
print(is_palindrome("mom"))
```
