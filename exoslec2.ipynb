```python
#excercice 1 
import math                     
def taylor(x, N):            
    result = 0                 
    for k in range(N):           
        result += x**k / math.factorial(k) 
    return result             
x = 1                            
N = 10                           
print(taylor(x, N))          
print(math.exp(x))            
```

```python
#excercice 4
def recaman(N):                  
    seq = [0]                   
    used = {0}                   
    for n in range(1, N):       
        candidate = seq[-1] - n  
        if candidate > 0 and candidate not in used:
            seq.append(candidate) 
        else:
            seq.append(seq[-1] + n) 
        used.add(seq[-1])        
    return seq                   
seq = recaman(50)                
plt.figure(figsize=(10, 4))     
for n in range(1, len(seq)):    
    x1, x2 = seq[n-1], seq[n]    
    center = (x1 + x2) / 2       
    radius = abs(x2 - x1) / 2    
    theta = np.linspace(0, np.pi, 100)  
    x_arc = center + radius * np.cos(theta)  
    y_arc = radius * np.sin(theta)          
    if n % 2 == 0:              
        y_arc = -y_arc           
    plt.plot(x_arc, y_arc, color="grey") 
plt.axis("off")                 
plt.title("Recaman Sequence")   
#plt.show()                       
```

```python
#excercice 5
import numpy as np
import matplotlib.pyplot as plt
theta = np.linspace(0, np.pi, 300)  
colors = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]
radii = np.linspace(1, 1.5, len(colors)) 

plt.figure(figsize=(8, 4))        

for r, c in zip(radii, colors):  
    x = r * np.cos(theta)        
    y = r * np.sin(theta)       
    plt.plot(x, y, color=c, linewidth=6)  

plt.axis("equal")                
plt.axis("off")                   
plt.title("The Rainbow of My life ")           
plt.show()                        
```

```python
#excercice 6
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def torus(R, r, u, v):
    x = (R + r * np.cos(v)) * np.cos(u)
    y = (R + r * np.cos(v)) * np.sin(u)
    z = r * np.sin(v)
    return x, y, z
u = np.linspace(0, 2 * np.pi, 1000)
v = np.linspace(0, 2 * np.pi, 1000)
u, v = np.meshgrid(u, v)
R = 2
r = 1  
x, y, z = torus(R, r, u, v)
fig = plt.figure(figsize=(15,7))
plt.subplots_adjust(hspace=0.5, top=0.5) 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='g', edgecolor='m', alpha=0.6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Torus')
plt.show()
```

```python
#excercice 7
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def model(x, a, b, c):
    return a * np.exp(-b * x) + c

np.random.seed(0)
x = np.linspace(0, 4, 50)
y_true = model(x, 2.5, 1.3, 0.5)
y = y_true + 0.2 * np.random.normal(size=len(x))

data = np.column_stack((x, y))
np.savetxt("fit_data.txt", data)

loaded_data = np.loadtxt("fit_data.txt")
x_data = loaded_data[:, 0]
y_data = loaded_data[:, 1]

params, covariance = curve_fit(model, x_data, y_data)
a_fit, b_fit, c_fit = params

x_fit = np.linspace(0, 4, 200)
y_fit = model(x_fit, a_fit, b_fit, c_fit)
plt.figure()
plt.scatter(x_data, y_data, label="Data")
plt.plot(x_fit, y_fit, label="Fitted curve")
plt.xlabel("x")
plt.ylabel("y")
plt.title("the exponential Curve Fitting")
plt.legend()
#plt.show()
```

```python
import turtle

sc = turtle.Screen()
pen = turtle.Turtle()

def semi_circle(col, rad, val):
    pen.color(col)
    pen.circle(rad, -180)
    pen.up()
    pen.setpos(val, 0)
    pen.down()
    pen.right(180)
col = ['violet', 'indigo', 'blue', 
       'green', 'yellow', 'orange', 'red']
sc.setup(600, 600)
sc.bgcolor('black')
pen.right(90)
pen.width(10)
pen.speed(4)
for i in range(7):
    semi_circle(col[i], 10*(
      i + 8), -10*(i + 1))
pen.hideturtle()
```

```python
#excercice 8
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import sympy as sp
def f(x):
    return x**2 + 10*np.sin(x)
result_num, error = integrate.quad(f, -10, 10)
x = np.linspace(-10, 10, 400)
y = f(x)
plt.figure()
plt.plot(x, y, label="f(x) = x^2 + 10 sin(x)")
plt.fill_between(x, y, alpha=0.3)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Integral of f(x)")
plt.legend()
#plt.show()
x_sym = sp.symbols('x')
f_sym = x_sym**2 + 10*sp.sin(x_sym)
result_sym = sp.integrate(f_sym, (x_sym, -10, 10))
print("Numerical integral (SciPy):", result_num)
print("Symbolic integral (SymPy):", float(result_sym))
```
